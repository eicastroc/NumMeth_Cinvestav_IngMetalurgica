```python
# cuadratic equation
def cuadratic(x, a, b, c):
    return a*(x)**2 + b*x + c
```

```python
fig, ax = plt.subplots(figsize=(6, 3))

x = np.linspace(0, 5, 100)
y = x**2

xi = np.array([2, 3, 4])
yi = xi**2

ax.plot(x, y)
ax.stem(xi, yi, basefmt='k:', linefmt='k:', markerfmt='or')

ax.set(xlim=(0, 5), ylim=(0.25),
       xticks=xi, xticklabels=[r"$x_{i-1}$", r"$x_{i}$", r"$x_{i+1}$"],
       yticks=yi, yticklabels=[r"$f(x_{i-1})$", r"$f(x_{i})$", r"$f(x_{i+1})$"],)
ax.grid(ls='--')
fig.tight_layout()

path = os.path.join("../", "images", "finiteDifferenceStencil.png")
fig.savefig(path, dpi=300)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 3))

x = np.linspace(0, 10, 5)
y = np.zeros_like(x)

xi = np.array([2, 4, 6, 8])
yi = np.zeros_like(xi)

ax.plot(x, y)
ax.plot(xi, yi, marker='o', ls='')

ax.set(xlim=(0, 10), ylim=(-1, 1),
       xlabel='x [cm]',
       yticks=[0], yticklabels=[""])
ax.grid(ls='--')

ax.annotate(r"$T(0)=100$", xy=(0, 0.125), ha='left', va='center')
ax.annotate(r"$x_{1}, T_{1}$", xy=(2, -0.125), ha='center', va='center')
ax.annotate(r"$x_{2}, T_{2}$", xy=(4, -0.125), ha='center', va='center')
ax.annotate(r"$x_{3}, T_{3}$", xy=(6, -0.125), ha='center', va='center')
ax.annotate(r"$x_{4}, T_{4}$", xy=(8, -0.125), ha='center', va='center')
ax.annotate(r"$T(10)=50$", xy=(10, 0.125), ha='right', va='center')

fig.tight_layout()

path = os.path.join("../", "images", "finiteDifference1DProblem.png")
fig.savefig(path, dpi=300)

plt.show()
```
