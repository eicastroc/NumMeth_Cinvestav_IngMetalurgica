## Problema:

**Problema:**


Dada una función real continua $f(x)$, que depende de la variable real $x$, se desa conocer la raíz (o las raíces) $x_r$ tal que: $f(x_r)=0$

**Ejemplo:**

Ecuación algebraica de tercer grado:

$$f(x) = x^3 + x^2 - 3x - 3$$


```python
def cubic(x, a, b, c, d):
    return a*(x)**3 + b*(x)**2 + c*x + d
```

```python
a, b, c, d = 1, 1, -3, -3
```

Esta ecuación tiene al menos una raíz, $x_r$, de tal forma que:

$$f(x_r) = x_r^3 + x_r^2 - 3x_r - 3 = 0$$

```python
fig, ax = plt.subplots()

x = np.linspace(-2, 2, 100)
y = cubic(x, a, b, c, d)

ax.plot(x, y)
ax.grid(ls='--', alpha=0.5)
ax.set(xlim=(-2, 2), ylim=(-5, 3), xlabel='x', ylabel='f(x)')
plt.show()
```

## ¿Cómo determinar la convergencia del método para encontrar raíces?

**Error relativo verdadero**

$$\varepsilon_t = \left|\frac{x_r - x_a}{x_r} \right|$$

$$\varepsilon_t (\%)= 100 \times \left|\frac{x_r - x_a}{x_r} \right|$$

donde:

- $x_r$: valor real (no siempre se conoce).

- $x_a$: valor aproximado (mediante un método numérico).

**Error relativo de aproximación:**

$$\varepsilon_a = \left| \frac{x_{i+1} - x_{i}}{x_{i+1}} \right|$$

$$\varepsilon_a (\%) = 100 \times \left| \frac{x_{i+1} - x_{i}}{x_{i+1}} \right|$$

donde:

- $x_{i+1}$: valor aproximado en la iteración actual.

- $x_{i}$: valor aproximado en la iteración previa.
