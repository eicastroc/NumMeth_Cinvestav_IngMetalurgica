```python
import os
import copy
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import bisect
```

```python
# cuadratic equation
def cuadratic(x, a, b, c):
    return a*(x)**2 + b*x + c
```

```python
def cubic(x, a, b, c, d):
    return a*(x)**3 + b*(x)**2 + c*x + d
```

```python
# x-domain and brackets for figures
x = np.linspace(-5, 5, 100)
xL = -2
xR = 2
```

# Método gráfico

```python
fig, axes = plt.subplots(nrows=1, ncols=4, 
                       sharex=True, sharey=True, figsize=(8, 2))

## cuadratic equation with no roots
a, b, c = 1, 0, 0.25
axes[0].plot(x, cuadratic(x, a, b, c), color='k')
#axes[0].scatter([xL, xR], [cuadratic(xL, a, b, c), cuadratic(xR, a, b, c)], color='b')

## cuadratic equation with two roots within brackets
a, b, c = -2, 0.5, 4
axes[1].plot(x, cuadratic(x, a, b, c), color='k')
#axes[1].scatter([xL, xR], [cuadratic(xL, a, b, c), cuadratic(xR, a, b, c)], color='b')
x0 = bisect(cuadratic, xL, 0, args=(a, b, c))
axes[1].scatter(x0, cuadratic(x0, a, b, c), marker='x', color='k')
x0 = bisect(cuadratic, 0, xR, args=(a, b, c))
axes[1].scatter(x0, cuadratic(x0, a, b, c), marker='x', color='k')


## cuadratic equation with one root within the brackets
a, b, c = 1, 2, -4
axes[2].plot(x, cuadratic(x, a, b, c), color='k')
#axes[2].scatter([xL, xR], [cuadratic(xL, a, b, c), cuadratic(xR, a, b, c)], color='b')
x0 = bisect(cuadratic, -5, xL, args=(a, b, c))
axes[2].scatter(x0, cuadratic(x0, a, b, c), marker='x', color='k')
x0 = bisect(cuadratic, xL, xR, args=(a, b, c))
axes[2].scatter(x0, cuadratic(x0, a, b, c), marker='x', color='k')



## cubic equation with three roots within brackets
a, b, c, d = -1, 0, 2, 0
axes[3].plot(x, cubic(x, a, b, c, d), color='k')
#axes[3].scatter([xL, xR], [cubic(xL, a, b, c, d), cubic(xR, a, b, c, d)], color='b')
for xl, xr in zip([-2, -0.5, 0.5], [-0.5, 0, 2]):
    x0 = bisect(cubic, xl, xr, args=(a, b, c, d))
    axes[3].scatter(x0, cubic(x0, a, b, c, d), marker='x', color='k')


## format for all graphs
for ax in axes:
    ax.axhline(0, ls='--', color='r')
    #ax.axvspan(xL, xR, color='b', alpha=0.1)
    #[ax.axvline(xx, ls='--', color='b') for xx in [xL, xR]]
    ax.set(xlim=(-5, 5), ylim=(-5, 5), xlabel=r'$x$', ylabel=r'$f(x)$')
    

fig.tight_layout()

path = os.path.join("../", "images", "graphical01.png")
fig.savefig(path,dpi=300)
```

# Ejemplos esquemáticos algorítmos acotados

## Método acotado

```python
fig, axes = plt.subplots(nrows=1, ncols=4, 
                       sharex=True, sharey=True, figsize=(8, 2))

## cuadratic equation with no roots
a, b, c = 1, 0, 0.25
axes[0].plot(x, cuadratic(x, a, b, c), color='k')
axes[0].scatter([xL, xR], [cuadratic(xL, a, b, c), cuadratic(xR, a, b, c)], color='b')

## cuadratic equation with two roots within brackets
a, b, c = -2, 0.5, 4
axes[1].plot(x, cuadratic(x, a, b, c), color='k')
axes[1].scatter([xL, xR], [cuadratic(xL, a, b, c), cuadratic(xR, a, b, c)], color='b')
x0 = bisect(cuadratic, xL, 0, args=(a, b, c))
axes[1].scatter(x0, cuadratic(x0, a, b, c), marker='x', color='k')
x0 = bisect(cuadratic, 0, xR, args=(a, b, c))
axes[1].scatter(x0, cuadratic(x0, a, b, c), marker='x', color='k')


## cuadratic equation with one root within the brackets
a, b, c = 1, 2, -4
axes[2].plot(x, cuadratic(x, a, b, c), color='k')
axes[2].scatter([xL, xR], [cuadratic(xL, a, b, c), cuadratic(xR, a, b, c)], color='b')
x0 = bisect(cuadratic, -5, xL, args=(a, b, c))
axes[2].scatter(x0, cuadratic(x0, a, b, c), marker='x', color='k')
x0 = bisect(cuadratic, xL, xR, args=(a, b, c))
axes[2].scatter(x0, cuadratic(x0, a, b, c), marker='x', color='k')



## cubic equation with three roots within brackets
a, b, c, d = -1, 0, 2, 0
axes[3].plot(x, cubic(x, a, b, c, d), color='k')
axes[3].scatter([xL, xR], [cubic(xL, a, b, c, d), cubic(xR, a, b, c, d)], color='b')
for xl, xr in zip([-2, -0.5, 0.5], [-0.5, 0, 2]):
    x0 = bisect(cubic, xl, xr, args=(a, b, c, d))
    axes[3].scatter(x0, cubic(x0, a, b, c, d), marker='x', color='k')


## format for all graphs
for ax in axes:
    ax.axhline(0, ls='--', color='r')
    ax.axvspan(xL, xR, color='b', alpha=0.1)
    [ax.axvline(xx, ls='--', color='b') for xx in [xL, xR]]
    ax.set(xlim=(-5, 5), ylim=(-5, 5), xlabel=r'$x$', ylabel=r'$f(x)$')
    

fig.tight_layout()

path = os.path.join("../", "images", "bracketing01.png")
fig.savefig(path,dpi=300)
```

# Ejemplos métodos no acotados

## Método de punto fijo

```python
fig, axes = plt.subplots(nrows=1, ncols=4, 
                       sharex=True, sharey=True, figsize=(8, 2))

## cuadratic equation with no roots
a, b, c = 1, 0, 0.25
axes[0].plot(x, cuadratic(x, a, b, c), color='k')
#axes[0].scatter([xL, xR], [cuadratic(xL, a, b, c), cuadratic(xR, a, b, c)], color='b')
ai, bi, ci = 1, 1, 0.25
xi = 0
axes[0].scatter(xi, cuadratic(xi, a, b, c), color='b')
for i in range(5):
    xi = cuadratic(xi, ai, bi, ci)
    #print(xi)
    axes[0].scatter(xi, cuadratic(xi, a, b, c), color='b', alpha=0.25)


## cuadratic equation with two roots within brackets
a, b, c = -2, 0.5, 4
f = lambda x: -2*x**2 + 0.5 * x + 4
#g = lambda x: -4*x**2 + 8
g = lambda x: np.sqrt(0.25*x + 2)
axes[1].plot(x, f(x), color='k')
x0 = bisect(f, xL, 0)
axes[1].scatter(x0, f(x0), marker='x', color='k')
x0 = bisect(f, 0, xR)
axes[1].scatter(x0, f(x0), marker='x', color='k')
# fixed point
xi = 0
axes[1].scatter(xi, f(xi), color='b')
for i in range(10):
    rel = 1.0
    xo = xi
    xi = g(x0)
    xi = (1-rel)*xo + rel*xi
    #print(xi)
    axes[1].scatter(xi, f(xi), color='b', alpha=0.25)


## cuadratic equation with one root within the brackets
a, b, c = 1, 2, -4
axes[2].plot(x, cuadratic(x, a, b, c), color='k')
#axes[2].scatter([xL, xR], [cuadratic(xL, a, b, c), cuadratic(xR, a, b, c)], color='b')
x0 = bisect(cuadratic, -5, xL, args=(a, b, c))
axes[2].scatter(x0, cuadratic(x0, a, b, c), marker='x', color='k')
x0 = bisect(cuadratic, xL, xR, args=(a, b, c))
axes[2].scatter(x0, cuadratic(x0, a, b, c), marker='x', color='k')
g = lambda x: np.sqrt(4 - 2*x)
ai, bi, ci = -0.5, 0, 2
xi = 0
axes[2].scatter(xi, cuadratic(xi, a, b, c), color='b')
for i in range(10):
    rel = 0.9
    #xi = rel * cuadratic(xi, ai, bi, ci)
    xi = rel * g(xi)
    #print(xi)
    axes[2].scatter(xi, cuadratic(xi, a, b, c), color='b', alpha=0.25)


## cubic equation with three roots within brackets
a, b, c, d = -1, 0, 2, 0
axes[3].plot(x, cubic(x, a, b, c, d), color='k')
#axes[3].scatter([xL, xR], [cubic(xL, a, b, c, d), cubic(xR, a, b, c, d)], color='b')
for xl, xr in zip([-2, -0.5, 0.5], [-0.5, 0, 2]):
    x0 = bisect(cubic, xl, xr, args=(a, b, c, d))
    axes[3].scatter(x0, cubic(x0, a, b, c, d), marker='x', color='k')
g = lambda x: np.power(2*x, 3/2)
xi = 0.5
axes[3].scatter(xi, cubic(xi, a, b, c, d), color='b')
for i in range(10):
    rel = 0.1
    #xi = rel * cuadratic(xi, ai, bi, ci)
    xi = rel * g(xi)
    #print(xi)
    axes[3].scatter(xi, cubic(xi, a, b, c, d), color='b', alpha=0.25)


## format for all graphs
for ax in axes:
    ax.axhline(0, ls='--', color='r')
    #ax.axvspan(xL, xR, color='b', alpha=0.1)
    #[ax.axvline(xx, ls='--', color='b') for xx in [xL, xR]]
    ax.set(xlim=(-5, 5), ylim=(-5, 5), xlabel=r'$x$', ylabel=r'$f(x)$')
    

fig.tight_layout()

path = os.path.join("../", "images", "fixedpoint01.png")
fig.savefig(path,dpi=300)
```

## Newton-Raphson

```python
f = lambda x: x**2 + 2*x - 4
fp = lambda x: 2**x + 2

x = np.linspace(0, 4, 100)
xi = 3
xi2 = xi - (f(xi)) / (fp(xi))
xi3 = xi2 - (f(xi2)) / (fp(xi2))
```

```python
fig, ax = plt.subplots(figsize=(4, 2))
ax.set(xlim=(0, 4), ylim=(-2.5, 12.5),
       xlabel=r"x", ylabel=r"f(x)")
ax.plot(x, f(x), c='k')
ax.axhline(0, ls=':', c='r')
ax.plot([xi, xi], [0, f(xi)], marker='.', ls='--', color='gray', alpha=0.5)
ax.plot([xi2, xi2], [0, f(xi2)], marker='.', ls='--', color='gray', alpha=0.5)

plt.arrow(x=xi, y=f(xi), dx=(xi2-xi), dy=-f(xi), shape='full', 
          color='b',
          length_includes_head=True, head_width=0.1, head_length=0.5)

plt.arrow(x=xi2, y=f(xi2), dx=(xi3-xi2), dy=-f(xi2), shape='full', 
          color='b',
          length_includes_head=True, head_width=0.1, head_length=0.5)

ax.annotate(r"$f(x_i)$", xy=(xi, f(xi)), va='top', ha='left')
ax.annotate(r"$f(x_{i+1})$", xy=(xi2, f(xi2)), va='bottom', ha='right')
ax.annotate(r"$f(x_{i+2})$", xy=(xi3, f(xi3)), va='bottom', ha='right')

ax.annotate(r"$0$", xy=(xi, 0), va='bottom', ha='left')
ax.annotate(r"$x_i$", xy=(xi, 0), va='top', ha='center')
ax.annotate(r"$x_{i+1}$", xy=(xi2, 0), va='top', ha='center')
ax.annotate(r"$x_{i+2}$", xy=(xi3, 0), va='top', ha='center')

path = os.path.join("../", "images", "newtonRaphson01.png")
fig.savefig(path,dpi=300)

fig.tight_layout()
plt.show()
```
