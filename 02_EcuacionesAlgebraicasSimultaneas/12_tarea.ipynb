**Aproximación gráfica**

Como puede verse, se esperaría encontrar cuatro pares de números que nos proporcionen las raíces del sistema.

```python
fig, ax = plt.subplots(figsize=(6, 4))

x = np.linspace(-6, 6, 61)
y = np.linspace(-6, 6, 61)

ax.plot(x, 4-2*x**2, label='f(x,y)', color='k', ls='-')
ax.plot(2-0.25*y**2, y, label='g(x,y)', color='k', ls='--')
ax.set(xlim=(-6, 6), ylim=(-6, 6),
       xlabel='x', ylabel='y')
ax.legend(loc='upper right')
ax.grid(ls="--")
plt.show()
```
