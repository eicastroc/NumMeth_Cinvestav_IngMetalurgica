## Problema

Dado un conjunto de puntos $(x_i, f(x_i))$ que representan una función continua en el intervalo $a \le x \le b$, evaluar la integral de dicha función en el propio intervalo.

$$ I = \int_{a}^{b} f(x) dx$$

Como se ilustra en el bloque de código debajo, la integral se puede interpretar como el area entre la curva de la función $f(x)$ y el eje-$x$.


```python
f = lambda x: np.cos(x)**2 + np.sin(x)
x = np.linspace(-2*np.pi, 2*np.pi, 101)

a, b = -np.pi, np.pi
ab = np.linspace(a, b, 101)

fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x, f(x))
ax.fill_between(ab, f(ab))
ax.axhline(0, ls='--', color='k')
ax.axvline(a, ls='--', color='r')
ax.axvline(b, ls='--', color='r')
plt.show()
```

Una estrategia general para evaluar numéricamente la integral, es expresarla como una suma discreta que aproxima el valor numérico de la misma:

$$ I = \int_{a}^{b} f(x) dx \approx \sum_{i=1}^{n} w_i f(x_i) + r_n$$

en donde $w_i$ son funciones de ponderación para $n$ evaluaciones de $f(x)$ en los puntos $x_i \in [a, b]$, y $r_n$ es el residual de la aproximación.

La selección del número de puntos de evaluación, $n$, de su localización en el intervalo $[a, b]$, y de las funciones de ponderación, $w_i$, tienen influencia en la precisión y la complejidad computacional para obtener una estimación.
