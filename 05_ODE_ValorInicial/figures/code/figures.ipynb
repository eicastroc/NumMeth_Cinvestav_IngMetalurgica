```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# cuadratic equation
def F(x):
    return -0.5*x**4 + 4*x**3 - 10*x**2 + 8.5*x + 1


def f(x):
    return -2*x**3 + 12*x**2 - 20*x + 8.5


def euler(f, x0, y0, dx):
    return y0 + f(x0)*dx
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

#
x = np.linspace(0, 4, 100)

a = 0
b = 0.5
dx = 0.125

ax.plot(x, F(x))
#ax.stem([a, b], [F(a), F(b)], basefmt='k:', linefmt='k:', markerfmt='None')

#ax.plot([0], F(0), ls='', marker='o')

x0,  y0 = 0, 1

X = []
for x0 in np.arange(0, 4, dx):
    ax.plot(x0+dx, euler(f, x0, y0, dx), ls='', marker='o', color='C1')
    y0 = euler(f, x0, y0, dx)

plt.show()
```
